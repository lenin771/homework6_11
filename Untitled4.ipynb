```python
from sympy.plotting import plot
from sympy import Symbol, solve, diff, sin, cos, sympify
from sympy.physics.control.control_plots import pole_zero_plot
import numpy as np
import matplotlib.pyplot as plt

# поиск корней функции
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
ans = solve(f,x)
print('Корни функции', ans)
```

```python
# Интервалы возрастания функции
diff_f = diff(f, x)
ans = solve(diff_f>0, x)
print('Интервал возрастания функции',ans)
```

```python
# Интервалы возрастания функции
diff_f = diff(f, x)
ans = solve(diff_f>0, x)
print('Интервал возрастания функции',ans)
```

```python
# Интервалы убывания функции
diff_f = diff(f, x)
ans = solve(diff_f<0, x)
print('Интервал убывания функции',ans)
```

```python
# График функции
x = np.linspace(0, 4 * np.pi, 100)

plt.figure()
plt.plot(x, np.sin(x)**2 - np.cos(x)**2 )
plt.show()
```

```python
# Вершины периодической функции
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
diff_f = diff(f, x)
print('Производная функции =',diff_f)
print('Вершины функции',solve(diff_f))
```

```python
# промежуток, где функция f>0
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
ans = solve(f>0, x)
print('Промежуток, где f>0',ans)
```

```python
# промежуток, где функция f<0
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
ans = solve(f<0, x)
print('Промежуток, где f<0',ans)
```
